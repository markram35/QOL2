### Note: this will work the same with thee standardized one

```python
standard = "yes" # <â€“ this is case sensitive
nstand = pd.read_csv("data_standardized.csv")
nstand.drop("Unnamed: 0", axis = 1, inplace = True)
```

```python
# nstand.keys()
```

```python
#initialization
newdf = pd.DataFrame()
#creates index
nstand_group = nstand.groupby("State Abbreviation")
newdf["State Abbreviation"] = nstand_group.groups.keys() 
#adds # of cities and renames it to "Number of Cities"
count = nstand.groupby("State Abbreviation")["City"].count()
newdf = newdf.merge(count,how = "inner", on = "State Abbreviation")
newdf = newdf.rename(columns = {"City":"Number of Cities"})
sumlist = ["Violent Crime Total","Property Crime Total","TotalPop","Men","Women","VotingAgeCitizen","Employed"]
#aggregates all non-character columns
for i in nstand.keys():
    if standard == "no":
        if nstand[i].dtype != "O" and i in sumlist:
            col = nstand.groupby("State Abbreviation")[i].sum()
            newdf = newdf.merge(col,how = "inner", on = "State Abbreviation")
            print("{} aggregated -- sum".format(i))
        if nstand[i].dtype != "O" and i not in sumlist:
            col = nstand.groupby("State Abbreviation")[i].mean()
            newdf = newdf.merge(col,how = "inner", on = "State Abbreviation")
            print("{} aggregated -- mean".format(i))
    if standard == "yes":
        if nstand[i].dtype != "O":
            col = nstand.groupby("State Abbreviation")[i].mean()
            newdf = newdf.merge(col,how = "inner", on = "State Abbreviation")
            print("{} aggregated".format(i))
```

```python
#save to csv
if standard == "yes":
    newdf.to_csv("State agg - standard.csv",sep=",")
    print("standardized csv downloaded")
#     agged.to_csv("State agg - standard.csv",sep=",")
#     print("standardized csv downloaded")
if standard == "no":
    newdf.to_csv("State agg - no standard.csv",sep=",")
    print("nonstandardized csv downloaded")
```
